```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
en_stopwords = stopwords.words("english")
```

```python
import string
string.punctuation
```

```python
punct_list = [char for char in string.punctuation]
```

```python
en_stopwords = punct_list + en_stopwords
```

```python
cleaned_response = []
for word in single_response:
    if word not in en_stopwords:
        # keep it
        cleaned_response.append(word)

cleaned_response
```

```python
cleaned_response = [word for word in single_response if word not in en_stopwords]
```

```python
from nltk.stem import PorterStemmer
from nltk.stem import LancasterStemmer
from nltk.stem import SnowballStemmer
```

```python
porter = PorterStemmer()
lancaster = LancasterStemmer()
snowball = SnowballStemmer("english")
```

```python
porter_stem = [porter.stem(word) for word in cleaned_response]
lancaster_stem = [lancaster.stem(word) for word in cleaned_response]
snowball_stem = [snowball.stem(word) for word in cleaned_response]
```

```python
stem_df = pd.DataFrame()
stem_df['original'] = cleaned_response
stem_df['porter'] = porter_stem
stem_df['lancaster'] = lancaster_stem
stem_df['snowball'] = snowball_stem

stem_df
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
stem_df['wordnet_lem'] = [wordnet_lemmatizer.lemmatize(word) for word in cleaned_response]

stem_df
```

```python
tagged_response = nltk.pos_tag(cleaned_response)
tagged_response
```

```python
[wordnet_lemmatizer.lemmatize(word[0], word[1]) for word in tagged_response]
```

```python
from nltk.corpus import wordnet
nltk.download('wordnet')
```

```python
def get_wordnet_pos(upenn_tag):
    if upenn_tag.startswith('J'):
        return wordnet.ADJ
    elif upenn_tag.startswith('V'):
        return wordnet.VERB
    elif upenn_tag.startswith('N'):
        return wordnet.NOUN
    elif upenn_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
get_wordnet_pos('MD')
```

```python
# creating a lemmatized word based on the correct pos
[wordnet_lemmatizer.lemmatize(word[0], pos=get_wordnet_pos(word[1])) for word in tagged_response]
```

```python
wordnet_lem_pos = [wordnet_lemmatizer.lemmatize(word[0], pos=get_wordnet_pos(word[1])) for word in tagged_response]
stem_df['wordnet_lem_pos'] = wordnet_lem_pos

stem_df
```
